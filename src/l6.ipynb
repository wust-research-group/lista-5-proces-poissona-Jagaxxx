## LISTA 6 - Procesy Poissona

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### METODA 1 - metoda generowania czas贸w oczekiwania na kolejny skok
gdzie: 

$I$ - liczba skok贸w $N_t$ na $[0, T]$, 

$S_i$ - momenty skok贸w, 

$t$ - suma czas贸w oczekiwania $T_i$

```python
def process_poiss_1(T, lambda_):
    S = []
    I = 0
    t = 0
    while True:
        U = np.random.uniform(0,1)
        t = t - 1/lambda_ * np.log(U)
        if t>T:
            break
        I = I + 1
        S.append(t)
    return S
```

```python
T = 20
lambda_ = 0.5
data = process_poiss_1(T, lambda_)
range_ = [i for i in range(len(data))]
plt.scatter(data, range_)
plt.title('Proces Poissona ver.1')
plt.xlabel('T')
plt.ylabel('N(t)')
plt.grid()
plt.show()
```

#### METODA 2 - generowanie moment贸w skoku

```python
def process_poiss_2(T, lambda_, size):
    n = np.random.poisson(lam=lambda_*T)
    U = []
    if n != 0:
        U = np.random.uniform(0, T, size)
        U = np.sort(U)
    return U
```

```python
size2 = 10
T2 = 20
lambda_2 = 0.5
range_2 = range(size2)
data2 = process_poiss_2(T = T2, lambda_ = lambda_2, size = size2)
plt.scatter(data2, range_2)
plt.title('Proces Poissona ver.2')
plt.xlabel('T')
plt.ylabel('N(t)')
plt.grid()
plt.show()
```
